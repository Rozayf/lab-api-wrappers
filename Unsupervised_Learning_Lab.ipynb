```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
pd.DataFrame(X_scaled,columns=X.columns).head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=1234)
kmeans.fit(X_scaled)
```

```python

clusters = kmeans.predict(X_scaled)

pd.Series(clusters).value_counts().sort_index()
```

```python
clusters
```

```python
import numpy as np
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
response.status_code 
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
title = soup.select("chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > cite")
```

```python
title = []
artist = []

num_iter = len(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite"))


title2 = soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite")
artist2 = soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em")

for i in range(num_iter):
    title.append(title2[i].get_text())
    artist.append(artist2[i].get_text())

print(title)
print(artist)
```

```python
Top100 = pd.DataFrame({"Title":title,
                       "Artist":artist,
                      })
```

```python
print('Please enter your favorite song')
```

```python
song = input('Please enter your favorite song ')

print('You Entered:', song)

if song in title:
   print('Great choice! Maybe you will also like this track:', Top100.sample(n=1))
else:
   print("Interesting taste, let me recommend a track according to your taste..")
```
